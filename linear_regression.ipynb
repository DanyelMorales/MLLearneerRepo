```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


X,Y =np.loadtxt("https://raw.githubusercontent.com/DanyelMorales/MLLearneerRepo/main/pizza.txt",skiprows=1, unpack=True)

def graph(X1,Y1):
  sns.set()
  plt.axis([0,50,0,50])
  plt.xticks(fontsize=15)
  plt.yticks(fontsize=15)
  plt.xlabel("Reservations", fontsize=30)
  plt.ylabel("Pizzas",fontsize=30)

  plt.plot(X1,Y1,"bo")
  plt.show()

graph(X,Y)
```

```python
def predict(X, w):
  return X * w

def loss(X,Y,w):
  error = predict(X,w)-Y
  squared_error = error ** 2
  return np.average(squared_error)

def train(iterations, X,Y, lr):
  w=0
  for i in range(iterations):
    tmpLoss=loss(X,Y,w)
    print(f"\n iteration={i} loss={tmpLoss}")
    if(loss(X,Y,w+lr)<tmpLoss):
      w+=lr
    elif(loss(X,Y,w-lr)<tmpLoss):
      w-=lr
    else:
      return w
  raise Exception(f"cannot convey after {iterations} iterations")

w=train(10000, X,Y, 0.01)
print(f"\nw={w}")
print(f"Prediction: x=20, Y=>{predict(20, w)}")
```
