```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import seaborn as sns

plt.ion()
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

X,Y =np.loadtxt("https://raw.githubusercontent.com/DanyelMorales/MLLearneerRepo/main/pizza.txt",skiprows=1, unpack=True)

ax.plot(X,Y, "bo")

def graph(X1,Y1, id):
    ax.set_xlim(0, id)
    ax.cla()
    ax.plot(X1,Y1, "bo")
    line1, = ax.plot(X1,X1, 'b-')
    display(fig)
    clear_output(wait = True)
    plt.pause(0.5)
    return line1


def updateLine(X1,Y1,theLine):
    theLine.set_ydata(Y1)
    theLine.set_xdata(X1)
    display(fig)
    clear_output(wait = True)
    #plt.pause(0.5)

```

```python
def showHistory(X,Y,historyW):
  theLine=graph(X,Y,1)
  for w in historyW:
    yHat=predict(X,w)
    updateLine(X,yHat,theLine)

def predict(X, w):
  return X * w

def loss(X,Y,w):
  error = predict(X,w)-Y
  squared_error = error ** 2
  return np.average(squared_error)

def train(iterations, X,Y, lr):
  w=0
  historyW=[]
  for i in range(iterations):
    historyW.append(w)
    tmpLoss=loss(X,Y,w)
    print(f"\n iteration={i} loss={tmpLoss}")
    if(loss(X,Y,w+lr)<tmpLoss):
      w+=lr
    elif(loss(X,Y,w-lr)<tmpLoss):
      w-=lr
    else:
      return [w,historyW]
  raise Exception(f"cannot convey after {iterations} iterations")

w, history =train(20000, X,Y, 0.01)
print(f"\nw={w}")
print(f"Prediction: x=20, Y=>{predict(20, w)}")
```

```python
showHistory(X,Y, history)
```
